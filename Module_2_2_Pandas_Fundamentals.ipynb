## Missing Values & Duplicates

**Simulate Missing Data:** Let's create a DataFrame with some simulated missing values to practice.

```python
import numpy as np
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, 4, 5],
        'C': [1, 2, 3, 4, np.nan]}
df_missing = pd.DataFrame(data)
print("Original DataFrame with missing values:")
print(df_missing)
```

```python
# To count missing values per column
df_missing.isnull().sum()
```

```python
df_dropped=df_missing.dropna()
print("Dataframe after dropping all the missing values")
print(df_dropped)
```

```python
df_missing.fillna({'A':df_missing['A'].mean()},inplace=True)
df_missing.fillna({'B':df_missing['B'].median()},inplace=True)
df_missing.fillna({'C':0},inplace=True)
```

```python
print(df_missing)
```

```python
data_dup = {'ID': [1, 2, 3, 2, 4, 1],
            'Value': ['A', 'B', 'C', 'B', 'D', 'A']}
df_duplicates = pd.DataFrame(data_dup)
print("\nOriginal DataFrame with duplicates:")
print(df_duplicates)
```

```python
# To print Duplicates
df_duplicates[df_duplicates.duplicated()]
```

```python
df_no_duplicates=df_duplicates[~df_duplicates.duplicated()]
df_no_duplicates
```
